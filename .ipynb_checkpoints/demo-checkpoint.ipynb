```python
import plotly.plotly as py
from plotly.graph_objs import *
import pandas as pd
import matplotlib.pyplot as plt

file_name = "./result/" + "result_1489729322.csv"
df = pd.read_csv(file_name)
x = df['round'].values.tolist()
y = df['count'].values.tolist()

results = y

maximum = max(results)
minimum = min(results)
mean = sum(results)*(1.0)/len(results)
medium = sorted(results)[len(results)/2]

print 'max = ', maximum
print 'min = ', minimum
print 'medium = ', medium
print 'mean = ', mean
plt.plot(x,y)
plt.show()
```
